# Finger 1: Análisis de datos de eventos en Trocafone

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import squarify
```

```python
#df = pd.read_csv('..\\\\Data\\tp1\\events.csv')
df = pd.read_csv('..\\Data\\tp1\\events.csv', low_memory=False)
```

## 1) - Analizar cuál es el tipo de evento predominante en el set de datos

```python
rankingEventos = df.groupby('event').size().sort_values(ascending=False).to_frame(name='cantidad').reset_index()
```

```python

# parámetros del treemap
x = 0.
y = 0.
#width = 200.
#height = 100.

# colores
cmap = matplotlib.cm.viridis
mini, maxi = rankingEventos.cantidad.min(), rankingEventos.cantidad.max()
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in rankingEventos.cantidad]


# labels
labels = ["%s\n%d" % (label) for label in zip(rankingEventos.event, rankingEventos.cantidad)]

# plot treemap
fig = plt.figure(figsize=(12, 10))
fig.suptitle("Trocafone - Web Analytics", fontsize=20)
ax = fig.add_subplot(111, aspect="equal")
ax = squarify.plot(rankingEventos.cantidad, color=colors, label=labels, ax=ax, alpha=.6 )
ax.set_xticks([])
ax.set_yticks([])
ax.set_title("Cantidad de ventos por tipo de evento\n", fontsize=14)


# Barra de escala
fig.text(.76, .9, "Escala", fontsize=14)
img = plt.imshow([rankingEventos.cantidad], cmap=cmap, alpha=.6)
img.set_visible(False)
fig.colorbar(img, orientation="vertical", shrink=.98)

# Footer
fig.text(.5, 0.1, "Total de eventos registrados: %d" % (rankingEventos.cantidad.sum()), fontsize=14, ha="center")
fig.text(.5, 0.07, "Análisis de usuarios que visitaron www.trocafone.com", fontsize=14, ha="center")

plt.show()
```

## 2) - Calcular cuales son los 5 productos (identificados por sku) con mayor cantidad de conversiones.

```python
top5sku=df[df['event'] == 'conversion'].groupby('sku').size().sort_values(ascending=False).to_frame(name='cantidad').reset_index().head(5)
```

```python
valores = top5sku.cantidad
labels = top5sku.sku

# plot

fig = plt.figure(figsize=(12, 10))
fig.suptitle("Trocafone - Web Analytics", fontsize=20)
ax = fig.add_subplot(111)
ax.set_title("Cantidad de ventos por tipo de evento\n", fontsize=14)
ax.barh(top5sku.sku, top5sku.cantidad)
plt.ylabel('SKU' , fontsize=16)
plt.xlabel('Cantidad de eventos' , fontsize=16)

#ax.set_xticks([])
#ax.set_yticks([])


```

# 3) - Averiguar cuáles son los 10 usuarios que registraron más actividad en la plataforma (es decir, realizaron más eventos).

```python
top10Person = df.groupby('person').size().sort_values(ascending=False).head(10).to_frame(name='cantidad').reset_index()
```

```python
valores = top10Person.cantidad
labels = top10Person.person

# plot

fig = plt.figure(figsize=(12, 10))
fig.suptitle("Trocafone - Web Analytics", fontsize=20)
ax = fig.add_subplot(111)
ax.set_title("Usuario más activos\n", fontsize=14)
ax.barh(top10Person.person, top10Person.cantidad)
plt.ylabel('Person' , fontsize=16)
plt.xlabel('Cantidad de eventos' , fontsize=16)

# Footer
fig.text(.5, -.004, "Total de eventos registrados: %d" % (rankingEventos.cantidad.sum()), fontsize=14, ha="center")
fig.text(.5, -0.045, "Análisis de usuarios que visitaron www.trocafone.com", fontsize=14, ha="center")

```
